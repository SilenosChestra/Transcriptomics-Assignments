```python
# Подгрузка необходимых библиотек

import scanpy as sc
import muon as mu
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scanpy.external as sce
import anndata2ri
anndata2ri.activate()
%load_ext rpy2.ipython
```

Разбираемся с данными. Ввиду разных неожиданных(нет) препятствий со стороны R в перемещении мультиомного .h5ad файла в Seurat, я решил сделать трюк и сначала создал объект muon, выполнил весь препроцессинг там, сохранил его в .h5mu, а далее с помощью библиотеки R прочитал его. Другие пути оказались обреченными на провал

```python
# Вспомогательная функция, чтобы создать объект muon (разделить модальности)
def create_MuData(adata, var_key="feature_types", layer="counts", batch_key="batch",
                  obs_columns=["cell_type", "batch"], batches=["s1d1", "s2d4"]):
  mdata = {}
  for mod in set(adata.var[var_key]):
    if layer == "X":
      X = adata[np.isin(adata.obs[batch_key], batches), adata.var[var_key] == mod].X
    else:
      X = adata[np.isin(adata.obs[batch_key], batches), adata.var[var_key] == mod].layers[layer]
    mdata[mod] = sc.AnnData(X=X, obs=adata[np.isin(adata.obs[batch_key], batches)].obs[obs_columns],
                            var=pd.DataFrame(index=f"{mod}_" + adata.var_names[adata.var[var_key] == mod]))
    mdata[mod].var_names_make_unique()
  mdata = mu.MuData(mdata)
  return mdata

adata_cite = sc.read_h5ad("GSE194122_openproblems_neurips2021_cite_BMMC_processed.h5ad")
mdata = create_MuData(adata_cite)
mdata["GEX"].layers["counts"] = mdata["GEX"].X.copy()
mdata["ADT"].layers["counts"] = mdata["ADT"].X.copy()

del adata_cite
```

Сначала скорректируем батч-эффект в двух модальностях отдельно (с помощью Harmony, а не Seurat CCA, чтобы достичь чистоты вводных для эксперимента)

Коррекция батч-эффекта в **GEX-модальности**

```python
# Убедимся, что в `adata.X` лежат каунты
mdata["GEX"].X = mdata["GEX"].layers["counts"].copy()

# Нормализация
sc.pp.normalize_total(mdata["GEX"], target_sum=1e4)
sc.pp.log1p(mdata["GEX"])

# Выделение HVG
sc.pp.highly_variable_genes(mdata["GEX"], n_top_genes=3000,
                            flavor="seurat_v3", layer="counts",
                            batch_key="batch")
# PCA и Harmony
mdata["GEX"].raw = mdata["GEX"]
sc.pp.scale(mdata["GEX"], max_value=10)
sc.tl.pca(mdata["GEX"], n_comps=30)
sce.pp.harmony_integrate(mdata["GEX"], key="batch", max_iter_harmony=20)

# kNN-граф
sc.pp.neighbors(mdata["GEX"], n_pcs=30, n_neighbors=20, use_rep="X_pca_harmony")

# UMAP
sc.tl.umap(mdata["GEX"])
```

Коррекция батч-эффекта в **ADT-модальности**

```python
# Убедимся, что в `adata.X` лежат каунты
mdata["ADT"].X = mdata["ADT"].layers["counts"].copy()

# CLR!!!
mu.prot.pp.clr(mdata["ADT"])

# Все наши переменные считаем HVG
mdata["ADT"].var["highly_variable"] = True

# PCA
mdata["ADT"].raw = mdata["ADT"] # запоминаем сырые экспрессии (важно)
sc.pp.scale(mdata["ADT"], max_value=10)
sc.tl.pca(mdata["ADT"])
sce.pp.harmony_integrate(mdata["ADT"], key="batch", max_iter_harmony=20)

# kNN-граф
sc.pp.neighbors(mdata["ADT"], n_pcs=30, n_neighbors=20, use_rep="X_pca_harmony")

# UMAP
sc.tl.umap(mdata["ADT"])
```

L2-нормализация (как по дефолту в Seurat) и построение новых графов соседей на нормированных PCs 

```python
# L2-нормализация PCA
mu.pp.l2norm(mdata, rep="X_pca_harmony")

# Построение графа соседей на (!) L2-отнормированных PCs
sc.pp.neighbors(mdata["GEX"], n_neighbors=20, n_pcs=30, use_rep="X_pca_harmony")
sc.pp.neighbors(mdata["ADT"], n_neighbors=20, n_pcs=30, use_rep="X_pca_harmony")
```

Апдейтим информацию и сохраняем наш muon объект в .h5mu

```python
mdata.update()
mdata.write("mdata.h5mu")
```

А теперь смертельный номе**R**. Обращаемся к R, где читаем наш .h5mu файл, делаем WNN, и строим UMAP на построенном графе, получая желанный WNN UMAP

```python
%%R

library(Seurat)
library(MuDataSeurat)
library(cowplot)

mdata <- ReadH5MU("mdata.h5mu")

# аналог mu.pp.neighbors(mdata)
mdata <- FindMultiModalNeighbors(
  mdata, reduction.list = list("ADTPCA_HARMONY", "GEXPCA_HARMONY"), 
  dims.list = list(1:30, 1:30), modality.weight.name = "RNA.weight"
)

# делаем umap на графах WNN
mdata <- RunUMAP(mdata, nn.name = "weighted.nn", reduction.name = "wnn.umap", reduction.key = "wnnUMAP_")

# дополнительно сделаем leiden (algorithm = 4, и есть leiden)
mdata <- FindClusters(mdata, graph.name = "wsnn", algorithm = 4, resolution = 2, verbose = FALSE)
```

Отрисуем наш umap на кластерах leiden и клеточных типах, и сравним с тем, одинаковы ли результаты с muon:
