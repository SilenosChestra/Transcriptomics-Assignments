Сначала посмотрим на распределение данных до трансформаций

```python
import scanpy as sc
import seaborn as sns
import scvelo as scv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# нужно, чтобы закинуть в curve_fit
def f_NB(x, a):
    return x + a * x**2

def mean_var_plot(adata, fit_NB=False, fit_poly=False, log=True,
                  xlim=None, ylim=None):
    
    # фитит прямую под функцию (в нашем случае должен найти нашу а (параметр тета)) 
    from scipy.optimize import curve_fit
    
    #отсекаем гены, у которых дисперсия > 0, ибо построение на log-шкале
    variances = np.var(adata.X.A, axis=0)
    means = adata.X.mean(axis=0).A[0][variances > 0] 
    variances = variances[variances > 0]
    
    # для красоты
    lim_min = min(np.min(means), np.min(variances))
    lim_max = max(np.max(means), np.max(variances))
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.plot([lim_min - 0.5, lim_max + 0.5], [lim_min - 0.5, lim_max + 0.5],
          linewidth=1, color="grey", label="$Var(E) = E$")
    sns.scatterplot(x=means, y=variances, ax=ax, linewidth=0, s=3)
    
    # ФИТИМ NEGATIVE BINOMIAL ПОД НАШИ ДАННЫЕ
    if fit_NB:
        popt_NB, _ = curve_fit(f_NB, means, variances) #variences тут - ydata, means - xdata
        fit_line_NB = np.array(range(int(lim_max))) * 0.01
        ax.plot(fit_line_NB, f_NB(fit_line_NB, *popt_NB), color="red", 
                label="$Var(E) = E + ${:.2f}$E^2$".format(popt_NB[0]))
    ax.set_ylabel("Variance")
    ax.set_xlabel("Mean")
    if log:
        ax.set_xscale("log")
        ax.set_yscale("log")
    if not(xlim is None):
        ax.set_xlim(left=xlim[0], right=xlim[1])
    else:
        ax.set_xlim(left=0, right=lim_max)
    if not(ylim is None):
        ax.set_ylim(bottom=ylim[0], top=ylim[1])
    else:
        ax.set_ylim(bottom=0, top=lim_max)
    ax.legend()

sns.set_style("ticks")
sc.set_figure_params(dpi=150)
```

```python
adata_smart = sc.read_h5ad("Smart-seq2.h5ad")
mean_var_plot(adata_smart, log=True)
```

В целом по графику видно, что овердисперсия начинается очень рано. Даже у низкоэкспрессированных генов в среднем дисперсия УЖЕ больше, чем среднее. Распределение не подходит под Пуассона точно.

```python
# первый метод - просто CP10K (должен убрать проблему глубины секвенирования библиотеки)
adata = adata_smart.copy()

sc.pp.normalize_total(adata, target_sum=1e4)
mean_var_plot(adata, log=True)
```

Интересно, что порядки после СР10К-трансформации снизились. Видимо, слухи не врут - в SMART-seq действительно большое количество UMI на клетку!

```python
# CP10K + логарифмирование
adata = adata_smart.copy()
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
mean_var_plot(adata, log=True)
```

Порядки еще больше уменьшились, и, судя по всему, мы избавились от овердисперсии, получив при этом андердисперсию. Можно ли говорить, что это успех? Да, еще осталось зависимость среднего от дисперсии (наша т.н. техническая дисперсия).

## Задание 2

Постройте постройте аналогичный заданию из занятия боксплот с распределением квардарат коэффециента корреляции между экспрессией гена и глубиной секвенирования клетки, только добавьте туда также и корреляцию для остатков регрессии SCTransform (и Analytic Pearson residuals). Есть ли при таких трансформациях зависимость между глубиной секвенирования клетки и экспрессией генов?

```python
adata_10x = sc.read_h5ad("PBMC_Satija.h5ad") #содержит несколько батчей
adata_10x = adata_10x[adata_10x.obs.donor == "P1"] # вытащим один конкретный батч
```

Сначала найдем в наших данных **Analytical Pearson residuals**

```python
adata_pr = adata_10x.copy()
adata_pr.layers['counts'] = adata_pr.X.copy()
sc.experimental.pp.normalize_pearson_residuals(adata_pr, theta=7.7)
# избавляемся от нулей,nan, статичных генов без дисперсии
mask = np.var(adata_pr.X, axis = 0) > 0
adata_pr = adata_pr[:, mask]
n_UMIs_pr = adata_pr.layers['counts'].sum(axis=1)
```

Теперь применим **SCTransform**

```python
import rpy2.robjects as ro
import anndata2ri
anndata2ri.activate()
%load_ext rpy2.ipython

adata_SCT = adata_10x.copy()
ro.globalenv["adata_counts"] = adata_10x
ro.r('SCT.results <- sctransform::vst(assay(adata_counts), vst.flavor="v2")')
```

```python
var_SCT = ro.r("SCT.results$gene_attr") # для каждого гена его параметры (удалилось около 3к генов)
SCT = ro.r("SCT.results$y") # остатки NB-регрессии

adata_SCT = adata_SCT[:, var_SCT.index] # фильтруем по генам
adata_SCT.layers["SCTransform"] = SCT.T
n_UMIs_SCT = adata_SCT.X.sum(axis=1)
```

```python
from scipy.stats import pearsonr

adata = adata_smart.copy()
adata.obs["total_UMI"] = adata.X.A.sum(axis=1) #количество каунтов на каждую клетку (в смартсек их 384)
n_UMIs = np.array(adata.obs["total_UMI"])

adata_PF = sc.pp.normalize_total(adata, copy=True)
adata_log1pPF = sc.pp.log1p(adata_PF, copy=True)
adata_PFlog1pPF = sc.pp.normalize_total(adata_log1pPF, copy=True)


# для каждого гена мы высчитываем корреляцию его каунтов по каждой клетке после преобразования и до преобразования
results_pearson = {
    "PF" : [pearsonr(adata_PF[:, gene].X.A.T[0], 
                     n_UMIs)[0] ** 2 for gene in adata.var_names],
    "log1pPF" : [pearsonr(
        adata_log1pPF[:, gene].X.A.T[0], n_UMIs
    )[0] ** 2 for gene in adata.var_names],
    
    "PFlog1pPF" : [pearsonr(
        adata_PFlog1pPF[:, gene].X.A.T[0], n_UMIs
    )[0] ** 2 for gene in adata.var_names],
    
    "Analytical_PR" : [pearsonr(adata_pr[:, gene].X.T[0], n_UMIs_pr)[0] ** 2 for gene in adata_pr.var_names],
    "SCTransform": [pearsonr(adata_SCT[:, gene].X.A.T[0], n_UMIs_SCT)[0] ** 2 for gene in adata_SCT.var_names]
}

#построил вместо DF словарь, ибо в DF не все колонки будут равны по размеру (SCTransform, Analyrical_PR убирают гены)
```

```python
sorted_keys, sorted_vals = zip(*sorted(results_pearson.items()))
```

```python
import matplotlib.pyplot as plt
ax = sns.boxplot(data=sorted_vals, showfliers=False)
ax.set_ylabel("$\mathrm{r}^2$ with cell depth")
ax.grid(alpha=0.3)
plt.xticks(plt.xticks()[0], sorted_keys)
ax.tick_params(labelsize=8, rotation = 90)
```

В принципе, как было и видно на семинарском графике, остатки Пирсона отработали лучше всего и реально убрали фактор глубины секвенирования
