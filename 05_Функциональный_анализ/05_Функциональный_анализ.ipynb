```python
# Открываем матрицу каунтов

count_matrix = pd.read_csv("LUAD_counts.tsv", sep='\t', index_col=0)

count_matrix.iloc[
    np.array([list(count_matrix.index).count(gene) for gene in df.index]) == 1
].to_csv("LUAD_counts_norepet.tsv", sep="\t")

count_matrix = pd.read_csv("LUAD_counts_norepet.tsv", sep='\t', index_col=0)


# Определяем каждой колонке его лэйбл (полученный ранее) для дизайна эксперимента в deseq2

coldata = pd.DataFrame({"condition":labels}, index = df.columns.values)
```

```python
%%R -i coldata,count_matrix

library(DESeq2)

dds <- DESeqDataSetFromMatrix(countData = count_matrix,
                              colData = coldata,
                              design = ~ condition)

dds <- DESeq(dds, test="LRT", reduced = ~1)

# filtering minimum counts 
dds <- dds[rowSums(counts(dds, normalized = TRUE)) >= 100,]


res <- results(dds, contrast=c("condition", "low_immune", "immune"), alpha=0.05)

#проводим урезание по LFC
resLFC <- lfcShrink(dds, res = res, coef="condition_low_immune_vs_immune", type="apeglm")
write.csv(resLFC, "resLFC.csv", quote=FALSE) 
```

```python
%%R
# посмотрим, сколько генов up- и downregulated при разных типах получения результатов
summary(resLFC)
summary(res)
```

Ранжирование по logFC и logFDR

```python
resLFC = pd.read_csv("resLFC.csv", sep=',').rename(columns = {"Unnamed: 0":"gene"})
```

```python
# LFC
resLFC_1 = resLFC.sort_values("log2FoldChange", ascending=False)[['gene','log2FoldChange']]

gs_res_1 = gp.prerank(
    resLFC_1,
    gene_sets=signatures.copy(),
    min_size=3,
    permutation_num=1000,
    outdir=None,
)

gs_res_1.res2d.head(10)
```

```python
from gseapy import gseaplot, heatmap
# отрисовка расчёта ES 
group = "Coactivation_molecules"
gseaplot(gs_res_1.ranking, term=group, **gs_res_1.results[group])
```

Видим, что те же Coactivation Molecules имеют больший по модулю Enrichment Score, чем в классическом GSEA (NES: = -0.85). Однако, т.к. наше ранжирование основывалось не на корреляции с фенотипом, а на изменении экспрессии между фенотипами, то можно предположить, что гены, принадлижащие к gene set Coactivation molecules, действительно оказались downregulated

```python
# —logFDR со знаком logFC

resLFC['signed_padj'] = -np.log(resLFC.padj) * np.sign(resLFC.log2FoldChange)
resLFC_2 = resLFC.sort_values("signed_padj", ascending=False)[['gene','signed_padj']]

gs_res_2 = gp.prerank(
    resLFC_2,
    gene_sets=signatures.copy(),
    min_size=3,
    permutation_num=1000,
    outdir=None,
)

gs_res_2.res2d.head(10)
```

```python
# отрисовка расчёта ES 
group = "Coactivation_molecules"
gseaplot(gs_res_1.ranking, term=group+" (LFC)", **gs_res_2.results[group])
gseaplot(gs_res_2.ranking, term=group+(" (padj_signed)"), **gs_res_2.results[group])
```

Для данной генной сигнатуры результаты примерно похожи на отфильтрованные по logFC

```python
group = "Proliferation_rate"
gseaplot(gs_res_1.ranking, term=group+" (LFC)", **gs_res_2.results[group])
gseaplot(gs_res_2.ranking, term=group+" (padj_signed)", **gs_res_2.results[group])
```
